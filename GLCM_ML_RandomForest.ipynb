```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
import os


"""
Classificador deve ser alterado conforme modelo a ser testado
"""

##########################################################################################

from sklearn.ensemble import RandomForestClassifier

##########################################################################################

label_dict = {'sound': 0,
              'wane': 1,
              'split': 2,
              'sound_knot': 3,
              'dry_knot': 4,
              'small_knot': 5,
              'core_stripe': 6,
              'mould': 7,
              'edge_knot': 8,
              'horn_knot': 9,
              'encased_knot': 10,
              'resin_pocket': 11,
              'UNKNOWN': 12,
              'leaf_knot': 13,
              'resin': 14,
              'bark_pocket': 15,
              'knot_hole': 16,
              'decayed_knot': 17,
              'moustache_knot': 18,
}


diretorio = 'Dataframes/GLCM_ML/'
arquivo = ['glcm1_ml',
           'glcm2_ml',
           'glcm3_ml',
           'glcm4_ml',
           'glcm5_ml',
           'glcm6_ml',
           'glcm7_ml',
           'glcm8_ml',
           'glcm9_ml',
           'glcm10_ml']

"""
Atenção para alterar o método, assim o nome do resultado e da pasta ficam corretos
"""
##########################################################################################

metodo = 'RandomForest'

if not os.path.exists('Dataframes/GLCM_ML/' + metodo):
    os.makedirs('Dataframes/GLCM_ML/' + metodo)

##########################################################################################


lista_acc = []
for indice in range(len(arquivo)):
    df_final = pd.read_pickle(diretorio + arquivo[indice])
    df_final.head()

    X = df_final.iloc[:,2: 8].values
    y = df_final.iloc[:,8:].values

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    """
    classifier muda conforme o modelo
    """
    ##########################################################################################

    classifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy', random_state = 0)

    ##########################################################################################

    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    matriz = confusion_matrix(y_test.argmax(axis=1),y_pred.argmax(axis=1),)
    matriz_normalizada = confusion_matrix(y_test.argmax(axis=1),y_pred.argmax(axis=1),normalize='true')
    df_resultado_n = pd.DataFrame(matriz_normalizada,columns=label_dict.keys(),index=label_dict.keys())
    df_resultado = pd.DataFrame(matriz,columns=label_dict.keys(),index=label_dict.keys())
    df_resultado.to_pickle(diretorio + metodo + "/" + arquivo[indice] + '_confusion_matrix')

    lista_acc.append(accuracy_score(y_test,y_pred))

    print(accuracy_score(y_test,y_pred))


"""
Cria o arquivo de texto com o Accuracy de cada modelo
"""

filename = 'Accuracy_' + metodo + '.txt'

if os.path.exists('Dataframes/GLCM_ML/' + metodo + '/' + filename):
    append_write = 'a'
else:
    append_write = 'w'
acc_writer = open('Dataframes/GLCM_ML/' + metodo + '/' + filename, append_write)
for i in range(len(arquivo)):
    acc_writer.write(str(arquivo[i]) + ": " + str(lista_acc[i]) + "\n")
acc_writer.close()

```
