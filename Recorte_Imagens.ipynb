```python
"""
Separar os setores gerados pelas coordenadas do arquivo manlabel.txt em imagens individuais.
As imagens originais estavam na pasta Imagens_Originais
A saída foi para uma pasta criada previamente chamada Imagens_Recortadas
Usando os nomes originais de referência da imagem como st1012
Retornando o nome da imagem_número do setor como st1012_50
"""
from PIL import Image
import pandas as pd

df = pd.read_csv('Dados_csv.csv')
nome_arquivo = df['image'].unique()
quantidade = df['image'].nunique()
for indice in range(quantidade):
    img = Image.open('Imagens_Originais/' + nome_arquivo[indice] + '.jpg')
    largura, altura = img.size
    df_image = df[df['image'] == nome_arquivo[indice]]
    df_quantidade = df_image['image'].count()
    for setor in range(df_quantidade):
        left = df_image.iloc[setor][2]
        upper = df_image.iloc[setor][1]
        right = df_image.iloc[setor][4]
        lower = df_image.iloc[setor][3]
        box = (left, upper, right, lower)
        img.crop(box).save('Imagens_Recortadas/' + str(nome_arquivo[indice]) + "_" + str(setor) + ".jpg" )
        arquivo = open('Recorte_log.txt',"a")
        arquivo.write(f'Imagem {nome_arquivo[indice]} extração {setor + 1}/{df_quantidade} \n')
        print(f'Imagem {nome_arquivo[indice]} extração {setor + 1}/{df_quantidade}')
arquivo.close()    
        
        


    
```
